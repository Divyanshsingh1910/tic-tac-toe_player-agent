```python
import random

random_integers = [random.randint(0, 4520) for _ in range(25)]
print(random_integers)
```

```python
import matplotlib.pyplot as plt


for i in random_integers:
    img = plt.imread(f"data_set/{i}.png")
    plt.imshow(img)
    plt.show()
```

```python
import os
# Create the data_set folder if it doesn't exist
if not os.path.exists("data_set_train"):
    os.makedirs("data_set_train")
```

```python
# import the module
import shutil

# Specify the path of the file you want to copy


# Specify the path of the destination directory you want to copy to
destination_directory = './data_set_train'

# Use the shutil.copy() method to copy the file to the destination directory
```

```python
for i in random_integers:
    file_to_copy = f"data_set/{i}.png"
    shutil.copy(file_to_copy, destination_directory)
```

```python
if not os.path.exists("data_set_test"):
    os.makedirs("data_set_test")

destination_directory = './data_set_test'
```

```python
for i in range(4520):
    if i not in random_integers:
        file_to_copy = f"data_set/{i}.png"
        shutil.copy(file_to_copy, destination_directory)
```
